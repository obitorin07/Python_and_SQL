# Task: Find the total order amount per country and display only the countries where the total order amount exceeds 700.
## Expected Output :

image.png

```python
def order_exceed(customers_df : pd.DataFrame , orders_df : pd.DataFrame)->pd.DataFrame:
  df =  pd.merge(customers_df ,orders_df , on='CustomerID', how ='inner')
  result = df.groupby('Country').agg(total_order_amount = ('OrderAmount','sum' )).reset_index()
  filter_result = result.loc[result.total_order_amount >=700]
  return filter_result

call_fun = order_exceed(customers_df, orders_df)
call_fun
```
